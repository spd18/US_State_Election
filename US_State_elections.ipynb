#ATTRIBUTE DESCRIPTION

**YEAR**: The year of the election (1976 in this dataset).

**STATE**: The full name of the state where the election took place (ARIZONA).

**STATE_PO**: The postal abbreviation of the state (AZ for Arizona).

**STATE_FIPS**: The Federal Information Processing Standard (FIPS) code for the state (4 for Arizona).
FIPS codes are unique numerical identifiers assigned to geographic areas.

**STATE_CEN**: Census Code for the state (86 for Arizona).
Used by the U.S. Census Bureau for statistical purposes.

**STATE_IC**: Internal Census Bureau code for the state (61 for Arizona).
Similar to state_cen, it’s an internal identifier.

**OFFICE**: The political office contested in the election (US SENATE in this dataset).
Other values in a broader dataset might include "US HOUSE," "GOVERNOR," etc.

**DISTRICT**: The congressional district for the election.
Since U.S. Senate races are state-wide, this value is always "statewide".

**STAGE**: The stage of the election, here "gen" (General Election).
Possible values in a broader dataset:
"gen" → General election
"pri" → Primary election
"runoff" → Runoff election

**SPECIAL** Indicates if the election was a special election (due to vacancy, etc.).
False means it was a regular election.

**CANDIDATE**: The name of the candidate running in the election.

**PARTY_DETAILED**: The detailed political party affiliation of the candidate, e.g., "REPUBLICAN", "DEMOCRAT", "LIBERTARIAN", "INDEPENDENT"

**WRITEIN**: True if the candidate was a write-in (not officially on the ballot).
False means they were officially listed on the ballot.

**MODE**: The method of voting recorded, always "total" in this dataset.
Could indicate specific voting methods (mail-in, early voting, etc.) in broader datasets.

**CANDIDATEVOTES**: The total number of votes received by the candidate.

**TOTALVOTES**: The total number of votes cast in the election for all candidates combined (741,210 in this case).

**UNOFFICIAL**: Indicates whether the election result is official or unofficial.
False means the results are official.

**VERSION**: A timestamp or version number for data tracking purposes.
"20210114" suggests the dataset was last updated on January 14, 2021.

**PARTY_SIMPLIFIED**: A simplified version of party_detailed, categorizing candidates into broader
political groups:
"REPUBLICAN" → Republican Party
"DEMOCRAT" → Democratic Party
"OTHER" → Independents or minor parties
"LIBERTARIAN" → Libertarian Party
